```python
import pandas as pd
import numpy as np
from statistics import mode
```

```python
df = pd.read_csv('iphone_purchase_records.csv')
df = df.drop(columns = ['Gender'],axis = 1)
df
```

```python
x = df.drop(columns = ['Purchase Iphone'],axis = 1)
y = df['Purchase Iphone']
```

```python
# euc_dist = lambda a1,b1,a2,b2: np.sqrt((a2-a1)**2 + (b2-b1)**2)

euc_dist = lambda pt1,pt2 : np.sqrt( (pt1[0] - pt2[0])**2 + (pt1[1] - pt2[1])**2 )


class KNN:
    def __init__(self,k):
        self.k = k

    def fit(self,xtrain,ytrain):
    
        self.xtrain = xtrain
        self.ytrain = ytrain
        

    def predict(self,xtest):
        
        majority_vote = []                                                               
        for i in range(len(xtest)):                                      
            dist_store = [( euc_dist(xtest.iloc[i],self.xtrain.iloc[j]),self.ytrain.iloc[j] )\
                            for j in range(len(self.xtrain))]
            dist_store = sorted(dist_store, key=lambda x: x[0])
            majority_vote.append( mode([d[1] for d,z in zip(dist_store,range(self.k))]) )
        
        return majority_vote
            
            
            
            
        
        
        
```

```python
test = KNN(3)
test.fit(x[:100],y[:100])
farris = test.predict(x[100:200])
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x[:100],y[:100])
real = neigh.predict(x[100:200])
```
