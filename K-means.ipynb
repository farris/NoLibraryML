```python
euclidean_distance = lambda p1,p2 : np.sqrt(    (p1[0]-p2[0])**2 + (p1[1]-p2[1])**2   )

class kmean:
    
    def __init__(self,k,iter_):
        
        self.k = k
        self.iter = iter_
        
    def predict(self,X):
        
        #intialize centroids randomly
        #closest points to those centroids get those labels
        #find mean of centroid
        
        initialize_c = [np.random.randint(0, high=500, size=1, dtype=int)[0] for _ in range(self.k)]
        centroids = [X[i,:] for i in initialize_c]
        
        for _ in range(self.iter):
            # find distance between each centroid and all other points
            dist_store = np.zeros((np.shape(X)[0],self.k)) 

            for idx,c in enumerate(centroids):

                dist_store[:,idx] = [euclidean_distance(c,X[j,:]) for j in range(len(X))]


            cluster_assingment = np.argmin(dist_store,axis=1)

            groups = [np.where(cluster_assingment == i)[0] for i in range(self.k)]

            centroids = [np.mean(X[g,:],axis=0) for g in groups]
            print(centroids)
            
            
            plt.scatter(X[groups[0],0],X[groups[0],1],color='r')
            plt.scatter(X[groups[1],0],X[groups[1],1],color='b')
            plt.scatter(X[groups[2],0],X[groups[2],1],color='g')
            plt.scatter(X[groups[3],0],X[groups[3],1],color='k')
            plt.scatter(X[groups[4],0],X[groups[4],1],color='y')
            plt.show()

        
        
        
        
        
        
        
```
