```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
import numpy as np

X, y = datasets.make_blobs(
    n_samples=150, n_features=2, centers=2, cluster_std=1.05, random_state=2
)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=123
)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.scatter(X_train[:, 0], X_train[:, 1], marker="o", c=y_train)
```

```python
class perceptron:
    
    def __init__(self,num_iter,alpha):
        self.num_iter = num_iter
        self.alpha = alpha
        
        
    def fit(self,x,y):
        '''   
        1. initialize weights
        2. move weights from incorrectly class
        3. We will use zero-one loss 
        3. grad((y - xw) \over w) = (y-(xw+b))(-x)
        
        '''
        w = np.zeros((np.shape(x)[1] ,1))
        bias = float(0)
        for _ in range(self.num_iter):
            for i in range(len(x)):
                activation = self._activation((x[i,:]@w)+bias)

                w = w + self.alpha*(y[i]-activation)*x[i,:][:,None]
                bias = bias + self.alpha*(y[i]-activation)

            plt.scatter(x[:, 0], x[:, 1], marker="o", c=y)
            plt.plot(np.linspace(-4,1,4),-1*w[0]/w[1]*np.linspace(-4,1,4)-(bias/w[1]))
            plt.show()
        
    def _activation(self,x):
        if x > 0: return 1
        else: return 0
        
```

```python
test = perceptron(100,.001)
test.fit(X_train,y_train)
```
